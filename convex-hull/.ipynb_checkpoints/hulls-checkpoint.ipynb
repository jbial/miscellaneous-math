# Convex Hull

If you've ever taken a course on optimization (or it may be called 'linear/nonlinear programming'), you might have heard of an abstract object called the ***convex hull***. To understand what a convex hull is, we need to define what *convexity* is. 

Convexity is an extremely rich and satisfying subject to study. We'll only be scratching the surface of it, by introducing the concept of a *convex set* - which is the only thing we'll need for this notebook. We consider sets $C \subseteq \mathbb{R}^2$ and $C \subseteq \mathbb{R}^3$. A set $C$ is convex if and only if the following property holds for any $\lambda \in \mathbb{R}$ such that $0 \leq \lambda \leq 1$:

$$\lambda \boldsymbol x + (1-\lambda)\boldsymbol y \in C\qquad \forall \boldsymbol x, \boldsymbol y \in C$$

Intuitively, this says that a set is convex if the line segment connecting any two points is also contained in the set. Here's a figure I stole from WolframAlpha:

![](https://mathworld.wolfram.com/images/eps-gif/ConvexConcave_1000.gif)

Now we can formally define what a convex hull is. Given a set an arbitrary set $S$, the convex hull is the smallest convex set, denoted $C = Conv(S)$, that contains $S$. That is, the smallest convex set C such that $S \subseteq C$.

A nice physical intuition of the convex hull is the shape left by wrapping a rubber band around an arbitrary shape.

We'll see later that the convex hull has many practical applications. For now, we'll focus on computing the convex hull for a given set of points in 2D.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def randset(n, dim=2, scale=1):
    """Generates a set of n gaussian random points
    
    Args:
        n (int): size of output set
        dim (int): dimensionality of the points
    """
    # randomly generate the parameters
    cov = np.random.rand(dim) * scale
    
    return np.random.multivariate_normal(np.zeros(dim), np.diag(cov), size=n)
```

```python
# generate random points
points = randset(200, dim=2, scale=10)
point = np.tanh(points)*np.sin(1/points)

plt.figure(figsize=(10, 10))
_ = plt.scatter(points[:, 0], points[:, 1])
```

## Algorithm 1: Gift Wrapping (Jarvis March)

The first algorithm for finding the convex hull that we'll talk about is called the *gift wrapping algorithm* or the Jarvis March. The basic idea is as follows:

----
1. Start at an extremal point i.e. left/right/top/bottom-most point (such a point is guaranteed to be in the convex hull)
2. Add initial extremal point to convex hull list
3. Pick another point to set as a candidate point
4. Repeat until candidate point == extremal point
 * Iterate through the points:
     * If a point is to the right/left/bottom/top of the line connecting the candidate and latest convex hull point:  
         * Set the candidate point as the selected point
 * Add candidate point to the convex hull
5. Return the convex hull
----
Notice that if there happens to be 3 collinear points in the set, then extra steps are required to identify if a point is a viable candidate point. In the following implementation, we will start with the right-most point which means we will be checking if candidate points lie to the *left* of the line formed between the previous candidate and a selected point.

The runtime of this particular algorithm is $O(|C||S|)$ making it an *output-sensitive algorithm* in that the runtime depends on the size of the solution itself. The value of $|C|$ could be much smaller than $|S|$, or in the worst case equal - leading to a runtime of $O(|S|^2)$.

In the implementation below, we use a simple cross-product test to check if a point is on the desired side of the candidate-selected point line. If the sign of the cross-product is positive then we accept the candidate, otherwise reject it.

```python
def cross_prod_magnitude(v1, v2):
    """Implements the magnitude of the cross product btwn 2 vectors 
    """
    return (v1[0] * v2[1] - v2[0] * v1[1])


def jarvis_march(S):
    """Finds the convex hull of a point set S
    """
    C = []
    
    # find the right-most point
    hull_pt = max(range(len(S)), key=lambda i: S[i][0])
    C.append(hull_pt)
    
    # initialize candidate point
    candidate = 0 if hull_pt != 0 else -1
    
    while (S[candidate] != S[C[0]]).all():

        # latest point on the convex hull
        latest = C[-1]
        latest_pt = S[latest]
        
        # reset candidate point and ensure it's not equal to the latest hull point
        candidate = (latest + (candidate == hull_pt)) % len(S)
        
        for i, pt in enumerate(S):
                
            xprod = cross_prod_magnitude(pt - latest_pt, S[candidate] - latest_pt)
            
            # correct side of the line OR if the candidate is on the hull itself
            # without first condition, candidate sometimes never gets updated
            if candidate == hull_pt or xprod > 0:
                candidate = i
            
        C.append(candidate)
        hull_pt = candidate
        
    # returns in counter-clockwise order
    return C
```

```python
jm_hull = jarvis_march(points)
```

```python
plt.figure(figsize=(10, 10))
plt.plot(*points[jm_hull].T, 'r--')
_ = plt.scatter(points[:, 0], points[:, 1])
```

## Algorithm 2: Graham Scan

The Graham scan convex hull algorithm works as follows:

----
1. Initalize a stack
2. Locate an extremal point (e.g. point with lowest y-coordinate)
3. Sort the point set by the angle the line created by each point with the extremal point makes with the x-axis
4. For each point:
    * If the preceding two points and the current point make a right=handed turn
      * pop the stack
    * push to the stack
5. Return the stack
----

This algorithm also introduces caveats if there are three collinear points in the sorted point list, but this case is simple to handle. Notice how the structure of the loop looks like a sliding window procedure where for each point we consider (point_(t-1), point_t, point(t+1)) to determine if they make a left/right hand turn. Thus, the runtime of the loop section is $O(|S|)$ and the runtime of the whole algorithm is dominated by the sorting step. The runtime of Graham scan algorithm is therefore $O(|S|\log|S|)$.


```python
def graham_scan(points):
    
    # get the lowest y-coord point in the set
    lowest = min(range(len(points)), key=lambda p: points[p][1])
    low_pt = points[lowest]
    
    # sort by angle & and send the extremal point the end of the list
    sorted_pts = sorted(
        range(len(points)),
        key=lambda p: 
            float('inf') if (points[p] == points[lowest]).all() 
            else -((points[p] - low_pt) / np.sqrt((points[p] - low_pt) @ (points[p] - low_pt)))[0]
    )
    
    # maintain a stack to store the convex hull
    C = [lowest]
        
    for p in sorted_pts:
        pt = points[p]
        
        # check if the lines made my by the current point and last two points make a right hand turn
        while len(C) > 1 and cross_prod_magnitude(pt - points[C[-1]], points[C[-2]] - points[C[-1]]) <= 0:
            C.pop()
        C.append(p)
        
    return C
```

```python
gs_hull = graham_scan(points)
```

```python
plt.figure(figsize=(10, 10))
plt.plot(*points[gs_hull].T, 'r--')
_ = plt.scatter(points[:, 0], points[:, 1])
```
